```python
# Parameters

NAME = []
AGE = []
OVA = []
POT = []
TEAM_NAME = []
CONTRACT = []
VALUE = []
WAGE = []
TOTAL_STATS = []
```

```python
pages = np.arange(0,541,60)
print(pages)
```

```python
url = ""
```

```python
for page in pages:
    page = requests.get("https://sofifa.com/players?offset="+str(page))
    
    soup=BeautifulSoup(page.content,'lxml')
soup.prettify()
```

```python

atri_name = []
for i in soup.findAll('div', attrs = {'class': 'bp3-text-overflow-ellipsis'}):
  
  atri_name.append(str(re.sub('^<div>.*">|\d{4}|[~]|\d{4}|"<.*?>"|Free|<span class="bp3-tag bp3-minimal bp3-intent-success">21</span>\xa0Diego Costa|<div class="bp3-text-overflow-ellipsis">|Jun 30,|<figure class="avatar avatar-sm transparent">|<img alt="" class="team" data-root="https://cdn.sofifa.com/teams/" data-src="https://cdn.sofifa.com/teams/\d+/\d+.png"|data-srcset="https://cdn.sofifa.com/teams/\d+/\d+.png 2x,|https://cdn.sofifa.com/teams/\d+/\d+.png 3x"|data-type="team" src="https://cdn.sofifa.com/teams/notfound_\d+.png"/>|<span class="bp3-tag bp3-minimal bp3-intent-success">On Loan</span>|<a href=".*|</figure>|</a><div class="sub">|\n|</div>','',str(i))))
```

```python
atri_name1 = [i for i in atri_name if i != '     ']
NAME =[i for i in atri_name1 if i != '']
```

```python
for i in soup.findAll('td', attrs = {'class': 'col col-ae'}):
 
  AGE.append(str( re.sub('<td class="col col-ae" data-col="ae">|</td>','',str(i))))
```

```python
for i in soup.findAll('td', attrs = {'class': 'col col-oa'}):
 
  OVA.append(str( re.sub('^<td>.*"|<td class="col col-oa" data-col="oa">|<span class="bp3-tag p p-\d{2}">|</span></td>','',str(i))))
```

```python
for i in soup.findAll('td', attrs = {'class': 'col col-pt'}):
 
  POT.append(str( re.sub('^<td>.*"|<td class="col col-pt" data-col="pt"><span class="bp3-tag p p-\d{2}">|</span></td>','',str(i))))
```

```python

find_team = []
for i in soup.findAll('div', attrs = {'class': 'bp3-text-overflow-ellipsis'}):
  find_team.append(str( re.sub('^<div>.*|\n\d{4} |[~]| \d{4}|\n|<a href=".*?" |rel="nofollow"|<div class="sub">|Free<a href="/players?na=7" | rel="nofollow"| title="Belgium">|<img alt="" class="flag" |data-src="https://cdn.sofifa.com/flags/be.png" data-srcset="https://cdn.sofifa.com/flags/be@2x.png 2x, https://cdn.sofifa.com/flags/be@3x.png 3x" src="https://cdn.sofifa.com/pixel.gif"/></a>|<span class="bp3-tag bp3-minimal bp3-intent-success">21</span>\xa0|<div class="bp3-text-overflow-ellipsis">|<span class="bp3-tag bp3-minimal bp3-intent-success">On Loan</span>|<figure class="avatar avatar-sm transparent">\n<img alt="" class="team" data-root="https://cdn.sofifa.com/teams/" data-src="https://cdn.sofifa.com/teams/.*?"| data-srcset="https://cdn.sofifa.com/teams/.*?"|<a href="/team/.*?">|</a><div class="sub">\nJun 30, 2022 |<span class="bp3-tag bp3-minimal bp3-intent-success">On Loan</span>\n| data-type="team" src="https://cdn.sofifa.com/teams/notfound_30.png"/>\n</figure>\n|</a><div class="sub">|\d{4} [~]\d{4}\n|</div>','',str(i))))
  
```

```python
TEAM_NAME1 = []
for i in find_team:
    if i not in TEAM_NAME1:
        TEAM_NAME1.append(i)
 
```

```python
 for i in TEAM_NAME1:
    if i not in NAME:
        TEAM_NAME.append(i)
 
```

```python
len(TEAM_NAME)
```

```python
cont_name = []
for i in soup.findAll('div', attrs = {'class': 'sub'}):
 
  cont_name.append(str( re.sub('^<div>.*"|<div class="sub">|\n|Free|' '|Jun 30, 2022|Jun 30, 2022 <span class="bp3-tag bp3-minimal bp3-intent-success">On Loan|<span class="bp3-tag bp3-minimal bp3-intent-success">|On Loan|</span></div>|</div>','',str(i))))
```

```python
cont_name2 = []
cont_name2 = [i for i in cont_name if i != ' ']
CONTRACT =[i for i in cont_name2 if i != '']
```

```python
for i in soup.findAll('td', attrs = {'class': 'col col-vl'}):
 
  VALUE.append(str( re.sub('^<td>.*"|<td class="col col-vl" data-col="vl">|M</td>|</td>','',str(i))))
```

```python
for i in soup.findAll('td', attrs = {'class': 'col col-wg'}):
 
  WAGE.append(str( re.sub('^<td>.*"|<td class="col col-wg" data-col="wg">|K</td>|</td>','',str(i))))
```

```python
for i in soup.findAll('span', attrs = {'class': 'bp3-tag p'}):
 
  TOTAL_STATS.append(str( re.sub('^<span>.*"|<span class="bp3-tag p">|</span>','',str(i))))
```

```python
pdname = pandas.DataFrame({"Name": NAME})
tname = pandas.DataFrame({"Team_Name": TEAM_NAME})
tcontract = pandas.DataFrame({"Contract_Year": CONTRACT,})
```

```python
df1 = pandas.DataFrame({"Age": AGE,
                   "Overall_Rating": OVA,
                   "Potential": POT})

df2 = pandas.DataFrame({"Value(M)": VALUE,
                   "Wage(K)": WAGE,
                   "Total_Stats": TOTAL_STATS})

```

```python
df = pandas.concat([pdname, df1, tname, tcontract, df2], axis=1)
```
